###Create a DataFrame with three columns: 'A', 'B', and 'C'.Fill the DataFrame with random numbers.

```python
df = pd.DataFrame(np.random.randint(0,100 , size=(5,3)),columns=['A','B','C'])
df
```

```python
colume_first = df['A'].mean()
colume_Two = df['B'].mean()
colume_Third = df['C'].mean()

print("mean for the first column",colume_first)
print("mean for the second column",colume_Two)
print("mean for the third column",colume_Third)
```

###Add a new column 'D' which is the sum of columns 'A' and 'B'

```python
df['D'] = df['A'] + df['B']
df
```

###Sort the DataFrame by column 'D' in descending order.

```python
sort_D = df.sort_values(by='D' ,ascending=False )
sort_D
```
